Get structural data of the airfoil

```python
#getting the info of max thickness and loc, max camber and loc from the webpage

import requests
from bs4 import BeautifulSoup
  
def structdetails(url):
 
    #open with GET method
    resp=requests.get(url)

    #http_respone 200 means OK status
    if resp.status_code==200:

        # we need a parser,Python built-in HTML parser is enough .
        soup=BeautifulSoup(resp.text,'html.parser')    
  
        # l is the list which contains all the text i.e news 
        l=soup.find("td",{"class":"cell1"})

        l=(str(l)) 

        l=l.replace("""<td class="cell1">""",'')

        l=l.split("<br/>")

        #print(l)

        name = l[1]
        #print("\nName:", name)
        
        Max_thickness = l[2].split(' ')[2]
        #print("\nMax thickness:", Max_thickness)

        Max_thickness_at = l[2].split(' ')[4]
        #print("\nMax thickness at:", Max_thickness_at)

        Max_camber = l[3].split(' ')[2]
        #print("\nMax camber:", Max_camber)

        Max_camber_at = l[3].split(' ')[4]
        #print("\nMax camber at:", Max_camber_at)    

        return [Max_thickness, Max_thickness_at, Max_camber, Max_camber_at] 
```

Download CSV files of all airfoils

```python
import os
import requests

filenames = []

def download(url: str, dest_folder: str):
    if not os.path.exists(dest_folder):
        os.makedirs(dest_folder)  # create folder if it does not exist

    filename = url.split('xf-')[-1] + '.csv'  # be careful with file names
    file_path = os.path.join(dest_folder, filename)
    
    r = requests.get(url, stream=True)

    if r.ok:
        print("saving to", os.path.abspath(file_path))
        with open(file_path, 'wb') as f:
            for chunk in r.iter_content(chunk_size=1024 * 8):
                if chunk:
                    f.write(chunk)
                    f.flush()
                    os.fsync(f.fileno())
        filenames.append(filename)

    else:  # HTTP status code 4XX/5XX
        print("Download failed: status code {}\n{}".format(r.status_code, r.text))
```
